# **Categorical diagrams - Distribution within categories**


1.   Boxplot
2.  Violinplot
3. Swarplot




```python
df1 = pd.read_csv('StudentsPerformance.csv')
df1.head()
```

## **Boxplot**

The boxplot displays the distribution of a continuous variable.

Quartiles separate the data into 4 equal numbers of data points:

25% of data points are in the lower quartile.

25% of data points are in the top quartile.

The 50th percentile (Q2) is the median.

EI - interquartile range (IQR) defines the width of the box.

50% of the data is in this box. (between Q1 and Q3).

"Whiskers" are defined by 1.5 x EI

outliers lie outside the whiskers.

```python
plt.figure(figsize = (10,4), dpi = 200)
sns.boxplot(data = df1, y = 'math score', x = 'parental level of education', hue = 'test preparation course')
plt.legend(bbox_to_anchor = (1.05, 0.5))
```

```python
df1.groupby(['parental level of education','test preparation course'	]).describe()['math score']
```

## **Violonplot**

The fiddle diagram plays a role similar to that of the box diagram.

It displays probability density across data using a KDE.

We can imagine it as a mirrored KDE diagram.

```python
plt.figure(figsize = (10,8), dpi = 200)
sns.violinplot(data = df1, x = 'reading score', y = 'parental level of education', hue = 'test preparation course', split = True)
plt.legend(bbox_to_anchor = (1.25, 0.5))
```

## **Swarmplot**

```python
plt.figure(figsize = (10,4), dpi = 200)
sns.swarmplot(data = df1, x = 'reading score', y = 'gender', hue = 'test preparation course', dodge = True)
plt.legend(bbox_to_anchor = (1.25, 0.5))
```

# **Comparison Diagram**
Comparison charts are essentially two-dimensional versions of previous charts.

The two main types of charts:


*   jointplot()
*   pairplot()


## **jointplot (Conjoint Diagram)**

```python
plt.figure(figsize = (10,10), dpi = 200)
sns.jointplot(data = df1, x = 'math score', y = 'reading score',  hue = 'gender')
```

## **pairplot**

Compare all column from Dataframe.
Work only on numeric column.

```python
sns.pairplot(df1)
```
